## 1. Fórmulas de Conversión

Para movernos entre estas dos escalas, utilizamos las siguientes relaciones matemáticas:

* **De Celsius (°C) a Fahrenheit (°F):** Multiplicamos los grados Celsius por 1.8 (o 9/5) y sumamos 32.
    $$°F = (°C \times 1.8) + 32$$

* **De Fahrenheit (°F) a Celsius (°C):** Restamos 32 a los grados Fahrenheit y el resultado lo dividimos entre 1.8.
    $$°C = \frac{°F - 32}{1.8}$$

---

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo 1: Convertir 25°C a Fahrenheit
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$
3.  **Multiplicar:** $25 \times 1.8 = 45$
4.  **Sumar:** $45 + 32 = 77$
* **Resultado:** 25°C equivalen a **77°F**.

### Ejemplo 2: Convertir 98.6°F a Celsius
1.  **Identificar la fórmula:** $°C = (98.6 - 32) / 1.8$
2.  **Restar:** $98.6 - 32 = 66.6$
3.  **Dividir:** $66.6 / 1.8 = 37$
* **Resultado:** 98.6°F equivalen a **37°C**.

## 3. Código en Python

Aquí tienes el programa estructurado con funciones, interactivo y fácil de leer:

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    c = float(input("\nIngresa los grados Celsius: "))
    f = (c * 1.8) + 32
    print(f"--- {c}°C es igual a {f:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    f = float(input("\nIngresa los grados Fahrenheit: "))
    c = (f - 32) / 1.8
    print(f"--- {f}°F es igual a {c:.2f}°C ---")

def menu():
    """Maneja el flujo principal del programa"""
    while True:
        print("\n--- CONVERSOR DE TEMPERATURA ---")
        print("1.- Convertir grados centígrados a fahrenheit")
        print("2.- Convertir grados fahrenheit a grados centígrados")
        print("3.- Salir")
        
        opcion = input("\nSelecciona una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida, intenta de nuevo.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```

### Explicación del Código:
* **`float()`**: Se usa para permitir números con decimales (como 36.5).
* **`f-strings`**: Usamos `f"{valor:.2f}"` para mostrar solo 2 decimales en el resultado final, lo que hace que se vea más limpio.
* **`while True`**: Crea un bucle infinito que solo se rompe cuando el usuario elige la opción 3 mediante la instrucción **`break`**.
* **Estructura**: Dividir el programa en funciones hace que el código sea organizado y profesional.

````python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    c = float(input("\nIngresa los grados Celsius: "))
    f = (c * 1.8) + 32
    print(f"--- {c}°C es igual a {f:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    f = float(input("\nIngresa los grados Fahrenheit: "))
    c = (f - 32) / 1.8
    print(f"--- {f}°F es igual a {c:.2f}°C ---")

def menu():
    """Maneja el flujo principal del programa"""
    while True:
        print("\n--- CONVERSOR DE TEMPERATURA ---")
        print("1.- Convertir grados centígrados a fahrenheit")
        print("2.- Convertir grados fahrenheit a grados centígrados")
        print("3.- Salir")

        opcion = input("\nSelecciona una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("Programa realizado por Caleb Dorado, NL = 14")
            break
        else:
            print("Opción no válida, intenta de nuevo.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
````
